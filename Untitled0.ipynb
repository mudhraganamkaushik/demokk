```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
d=pd.read_csv("Transformed_Housing_Data2.csv")
d.head()
```

```python
#scaling the data
from sklearn.preprocessing import StandardScaler
scaler =StandardScaler()
y=d['Sale_Price']
x=scaler.fit_transform(d.drop(columns=['Sale_Price']))
x=pd.DataFrame(data=x,columns=d.drop(columns=['Sale_Price']).columns)
x.head()
```

```python
#check and remove corel
x.corr()
```

```python
#to extract high pair of correlation
#by this we store the the corr which is 0.5
k=x.corr();
z=[[str(i),str(j)] for i in k.columns for j in k.columns if(k.loc[i,j]>abs(0.5))&(i!=j)]
z,len(z)
#by this we can see 32 corr which are greater than 0.5
```

```python
# remove multicorr by import vif
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif_data=x
#list compresion
VIF =pd.Series([variance_inflation_factor(vif_data.values,i)for i in range(vif_data.shape[1])],index=vif_data.columns)
VIF
#we see only 7 out of 16 ind var
```

```python
 #remove vif which are greater than 5and in this we remove which is max
 VIF[VIF ==VIF.max()].index[0]
```

```python
def MC_remover(data):
  vif=pd.Series([variance_inflation_factor(data.values,i)for i in range(data.shape[1])],index=data.columns)
  if vif.max() >5:
    print(vif[vif ==vif.max()].index[0],'has been removed')
    data=data.drop(columns=[vif[vif==vif.max()].index[0]])
    return data
  else:
    print('no multicolenear')
    return data
```

```python
for i in range(7):
  vif_data=MC_remover(vif_data)

vif_data.head()
```

```python
# above to remove multico linear
VIF =pd.Series([variance_inflation_factor(vif_data.values,i)for i in range(vif_data.shape[1])],index=vif_data.columns)
VIF, len(vif_data.columns)
```

```python
#removed 30 to 28
#training the model
x=vif_data
y=d['Sale_Price']
```

```python
from sklearn.model_selection import train_test_split
x_t,x_te,y_t,y_te=train_test_split(x,y,test_size=0.3,random_state=101)
x_t.shape,x_te.shape,y_t.shape,y_te.shape
```

```python
#implement the linear reg
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_t,y_t)
```

```python
lr.coef_#all coef of m0=0 etc m
```

```python
pred=lr.predict(x_te)
lr.score(x_te,y_te)
#this score is rsquare
```

```python
#verify the assumption
res=pred-y_te
res_tab=pd.DataFrame({'res':res,'pred':pred})
res_tab=res_tab.sort_values(by='pred')
z=[i for i in range(int (res_tab['pred'].max()))]
k=[0 for i in range(int (res_tab['pred'].max()))]
```

```python
plt.figure(dpi=130)
plt.scatter(res_tab['pred'],res_tab['res'],color='red',s=2)
plt.plot(z,k,color='green',linewidth=3,label='regression line')
plt.ylim(-800000,800000)
plt.show()
```

```python
#bar graph for check outliears
#model coefficients
coeff_tab=pd.DataFrame({
    'col':x_t.columns,
    'coefficients':lr.coef_
})
coeff_tab=coeff_tab.sort_values(by='coefficients')
#plot it with horizontal bar graph so that we can tell the relation between them
```
